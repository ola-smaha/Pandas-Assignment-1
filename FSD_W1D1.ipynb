## Inspecting the Data

A critical first step in data analysis, is inspecting the data. Several Pandas functions allow us to look at the different aspects of our dataset

### Heads and Tails

We can inspect the first 5 rows of the dataset using the following function:

```
df.head()
```
This function can also be used to inspect the first N rows of the dataset by doing the following:

```
N = 15
df.head(N)
```

The same operations can be used to inspect the last 5 or N rows of the dataset:

```
df.tail()
```

````python
df.head() # prints the first 5 rows of the dataset
````

````python
df.tail() # prints the last 5 rows of the dataset
````

````python
# Try it yourself:

N = 2 # number of rows we would like to display
df.head(N)
````

### Displaying Columns

In our case, not all columns are displayed when printing the first N rows. This can be fixed using the below code snippet:

```
with pd.option_context('display.max_rows', 5, 'display.max_columns', None):
    display(df)
```

By changing the numbers in front of each parameter we can control how many rows and columns we display. The use of **display** is required because pd.option_context settings only apply to **display**.

````python
# Try changing the parameters:
with pd.option_context('display.max_rows', 15, 'display.max_columns', None):
    display(df)
````

You can see all of the columns by using the below function:

```
df.columns
```

````python
df.columns
````

### Dataset Information

You might be tempted to know how many rows and columns there are in total in your dataset:

````python
df.shape
````

The first value is the number of rows, and the second, the number of columns. Building on this, and the previous 2 sections, you might be even more interested in the information contained within your dataset:

````python
df.info()
````

At first glance, it might be a bit intimidating looking at the output of this function, but it is simpler than it actually looks.


1. The first column we see '#' is the index of the column to its right.
2. The second column 'Column', contains the names of all the columns in the dataset.
3. The third column, 'Non-Null Count', contains the number of actual values in each column. If a column contains a null (empty) value, its corresponding number would be Number of Rows - Number of Missing Values.
4. The last column contains the data types of the various columns in our dataset.



### Selecting Columns

We can easily look at the values contained in each column by using the following notations:

```
df['column_name']
```
or
```
df.column_name
```

````python
df.head()
````

We can take a look at all the number of bedrooms for all the houses by writing the following

````python
df[['bedrooms']]
````

We can also specify the range we want to look at. For example, if we want to look at the first 5 houses

````python
start_index = 1
end_index = 5
df.bedrooms[start_index:end_index+1] ## Try using a different range here
````

````python
# What happens if we run the below command
df[df.bedrooms == 0]
````

We can also inspect two or more columns at the same time by writing the following

```
df[['column_name1', 'column_name2', 'column_name3'...]]
```

It is necessary to include the names within a double set of brackets: the first, outer set refers to the fact that we are selecting columns, the second, inner set of brackets refers to a list of column names we wish to display.

A simpler, and cleaner way of going about it, would be to create a list containing the names of the columns, and then passing it to our dataframe's brackets

```
list_columns = ['column_name1', 'column_name2', 'column_name3']
df[list_columns]
```

````python
list_columns = ['bedrooms', 'floors', 'price', 'waterfront']
df[list_columns]
````

Now that we have taken a quick glance at the dataset, we move on to each column, where we inspect its contents and try to make sense of the information it provides us with.


### Variable reference:

**id**: An identifier for each sale <br>

**sqft_living**: The living area of the house (apartment space) in square feet<br>
**sqft_lot**: The total area of the house in square feet<br>
**floors**: The number of floors <br>
**waterfront**: A variable indicating whether the house has a waterfront view; 0 if it doesn't, 1 if it does<br>

**condition**: An variable that takes the values from 1 to 5 representing the condition of the house<br>
**grade**: An variable that takes the values from 1 to 13 representing the grade given to the house<br>
**sqft_above**: How many square feet the house is above the ground level (so excluding basements and underground areas)<br>
**sqft_basement**: The square footage of the house's basement<br>
**yr_built**: The year the house was built<br>
**yr_renovated**: The year the house was renovated last<br>
**zipcode**: The area zipcode where the house is in<br>
**lat**: The latitude of the house's location<br>
**long**: The longitude of the house's location<br><br>
*Side note: Latitude and longitude are parameters or coordinates that help determine the location of any place on Earth. Both latitude and longitude are angles that uniquely define points on the Earth's sphere. They constitute a coordinate system or scheme to locate or identify geographic positions anywhere on the Earth's surface.*<br>

### Variable Types:

**Note**: The 64 in int64 refers to the memory allocated to hold this character. It represents signed integers with values that range from negative 9,223,372,036,854,775,808 through positive 9,223,372,036,854,775,807.

**Nominal Categorical**: id, zipcode *(as there is no order, and the numbers are unqiue identifiers)*<br>

#### Some insights:

````python
no_water_view = df[df['waterfront'] == 0]
water_view = df[df['waterfront'] == 1]
````

````python
print(water_view['price'].mean().round(2),no_water_view['price'].mean().round(2))
````

The average price of the houses with no waterfront view (531,653.41) is much smaller than that of houses with waterfront view (1,662,524.18). This potentially indicate a correlation between the two variables 'waterfront' and 'view'.

### Max and Min values:

Let's check the maximum and minimum values of some of our variables:

````python
df['bedrooms'].max()
````

````python
df[df['bedrooms']==33]
````

A house with 33 bedrooms has 1 bathroom other than the one exclusive to one of the bedrooms, something seems off...

````python
df.nlargest(5,'price')
````

The priciest house has 6 bedrooms, much less than 33. It has more bathrooms, a bigger area, a better view, and is graded higher. This could mean that there was a mistake while recording the data, that entry might be considered an outlier.

````python
df.nsmallest(10,'sqft_living')
````

If we look at the price, sqft_living, and the sqft_lot variables, we could notice some kind of relation between the three. So, one might affect the other; could be worth looking into and drawing a conclusion.
